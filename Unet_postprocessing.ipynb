```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the two images
image1 = cv2.imread('ground_truth.png')
image2 = cv2.imread('prediction.png')

# Resize images to the same dimensions if necessary
if image1.shape != image2.shape:
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Define transparency level (alpha value)
alpha = 0.8

# Overlay images with different colors and adjust transparency
overlay = cv2.addWeighted(image1, 1, image2, 1 - alpha, 1)

# Display the overlaid image
cv2_imshow(overlay)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save the overlaid image
cv2.imwrite('overlay_image.jpg', overlay)
```
